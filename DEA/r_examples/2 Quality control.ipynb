### MVA plot
График визуализирует различия между измерениями, выполненными в двух выборках, путем преобразования данных в шкалы M (логарифмический коэффициент) и A (среднее значение) с последующим нанесением этих значений на график.


Формула очень простая

$$M = log_2(E_1) - log_2(E_2)$$
$$A = \frac{log_2(E_1) + log_2(E_2)}{2}$$

где $$E_1- это первая экспрессия$$ , а $$E_2$$ - это вторая
То есть **M** - это логарифм отношения экспрессий, а **A** - это среднее значение логарифма двух экспрессий 

```r
# Выведем сравнение только первых двух сэмплов
MAplot(affyExpressionFS[, 1:2], pairs=TRUE)
```

### PCA

Анализ главных компонент

```r
eset_log <- exprs(affyExpressionFS)
eset_log <- log2(eset_log)
eset_log
```

```r
# Уменьшаем кол-во данных, потому что их слишком много))
edge <- nrow(eset_log) / 20
small_eset <- eset_log[1:edge, 1:ncol(eset_log)] 
nrow(small_eset)
```

```r
PCA_raw <- prcomp(t(small_eset), scale. = FALSE)
```

```r
se <- PCA_raw$sdev^2
percentVar <- round(100 * se / sum(se), 1)
percentVar
```

```r
dataGG <- data.frame(PC1 = PCA_raw$x[,1], PC2 = PCA_raw$x[,2])
dataGG
```

```r
ggplot(dataGG, aes(PC1, PC2)) +
geom_point()+
ggtitle("PCA plot of the log-transformed raw expression data") + 
xlab(paste0("PC1, VarExp: ", percentVar[1], "%")) + 
ylab(paste0("PC2, VarExp: ", percentVar[2], "%"))
```
