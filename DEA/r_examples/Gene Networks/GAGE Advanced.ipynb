```r
# Загрузим библиотеки
packages <- c("pathview",
              "gage",
              "gse16873",
              "kegg.gs")

if (!requireNamespace(packages, quietly = TRUE)) {
    BiocManager::install(packages)
}
```

```r
# Подключаем библиотеки
library(gage)
library(pathview)
data(gse16873)
data(kegg.gs)
```

```r
# Получаем индексы колонок двух классов
refSamples <- grep('HN', colnames(gse16873))
refSamples
compSamples <- grep('DCIS', colnames(gse16873))
compSamples
```

```r
# Посмотрим на данные
head(gse16873)
```

```r
# Проведем GSEA по KEGG
gse16873.kegg.p <- gage(gse16873,
                        gsets = kegg.gs,
                        ref = refSamples,
                        samp = compSamples)
```

```r
# Посмотрим на то, что получилось
head(gse16873.kegg.p$greater)
```

```r
# Отфильтруем данные по q.val 
sel <- gse16873.kegg.p$greater[, "q.val"] < 0.1 & !is.na(gse16873.kegg.p$greater[,"q.val"])
head(sel)
```
