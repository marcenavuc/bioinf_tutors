Выравнивание последовательностей и построение филогенетического дерева будем делать с помощью MUSCLE.
Перед началом необходимо скачать muscle3.8.31_i86win32.exe (http://www.drive5.com/muscle).

Необходисо выполнить в терминале

```
muscle3.8.31_i86win32.exe -in usflu.fasta -out aln.fasta -tree1 aln.nwk
```

````r
library(ggtree)
````

````r
#Считываем дерево
tree <- read.tree("aln.nwk")
#ggtree - строит дерево, geom_tiplab - подписывает имена листьям 
p <- ggtree(tree, branch.length = 'none') + geom_tiplab(size=2.25)
#msaplot - связывает дерево и файл с последовательностями
msaplot(p, "aln.fasta", offset=6, width=2)
````
