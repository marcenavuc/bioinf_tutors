```r
library(muscle)
```

```r
fasta_file <- readDNAStringSet("usflu.fasta", format="fasta", nrec=-1L, skip=0L, seek.first.rec=FALSE, use.names=TRUE)
muscle(stringset = fasta_file, tree1 = "aln.nwk", fastaout = "aln.fasta")
```

```r
library(ggtree)
```

```r
tree <- read.tree("aln.nwk")
p <- ggtree(tree, branch.length = 'none') + geom_tiplab(size=2.25)
msaplot(p, "aln.fasta", offset=6, width=2)
```
