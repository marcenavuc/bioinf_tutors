# ggtree для визуализации филогенетических деревьев

```r
if (!requireNamespace("ggtree", quietly = TRUE))
    BiocManager::install("ggtree")
library(ggtree)
```

```r
tree <- read.tree("phylotree.nwk")
p <- ggtree(tree, branch.length = 'none') + geom_tiplab(size=2.25)
msaplot(p, "aln-fasta.fasta", offset=6, width=2)
```

```r
ggsave("rendered_tree.pdf")
```
